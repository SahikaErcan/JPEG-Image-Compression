```python
# col'un 0 ile 255 arasında olduğundan emin olalım

def Clamp(col):  
    col = 255 if col > 255 else col
    col = 0 if col < 0 else col
    return int(col)
```

```python
#Y, Cr ve Cb'yi RGB renk uzayına dönüştürür

def ColorConversion(Y, Cr, Cb):
    R = Cr * (2 - 2 * 0.299) + Y
    B = Cb * (2 - 2 * 0.114) + Y
    G = (Y - 0.114 * B - 0.299 * R) / 0.587
    return (Clamp(R + 128), Clamp(G + 128), Clamp(B + 128))
```
